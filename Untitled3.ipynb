```python
import pandas as pd

# 1. 데이터 불러오기
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch6/diabetes_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch6/diabetes_test.csv")

print("===== 데이터 크기 =====")
print("Train Shape:", train.shape)
print("Test Shape:", test.shape)
print("\n")

# 2. 데이터 정보 및 결측치 확인
print("===== 데이터 정보(자료형) =====")
print(train.info())
print("\n")

print("===== 결측치 확인 =====")
print("Train:", train.isnull().sum().sum())
print("Test:", test.isnull().sum().sum())
print("\n")

# 3. Target 분리 (따옴표 수정)
print("===== target 빈도 =====")
print(train['Outcome'].value_counts())

target = train.pop('Outcome')

# 4. 데이터 분할
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

print("===== 분할된 데이터 크기 =====")
print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
print("\n")

# 5. 모델 학습 및 검증
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)

# 검증 데이터로 확률 예측 (ROC-AUC 등을 확인하고 싶을 때 사용)
val_pred = rf.predict_proba(X_val)

# 6. 테스트 데이터 예측 및 파일 저장
test_pred = rf.predict_proba(test)

# 제출용 데이터프레임 생성 (오타 수정: {'pred': test_pred[:, 1]})
submit = pd.DataFrame({'Outcome': test_pred[:, 1]})
submit.to_csv("result.csv", index=False)

# 7. 제출파일 확인
print("===== 제출파일 (샘플 5개) =====")
print(pd.read_csv("result.csv").head())
```
