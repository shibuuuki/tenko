```python
import pandas as pd
# df = pd.read_csv("air_quality.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/air_quality.csv")

# 1) IQR 계산
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3 - Q1

# 2) 상한 및 하한 계산
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

# 3) 이상치 식별
outliers = df[(df['CO2'] < lower) | (df['CO2'] > upper)]

# 4) 이상치 수
print(len(outliers))
```

```python
import pandas as pd

# 1. 데이터 로드 (URL 오타 수정: ovedim -> lovedlim)
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/loan.csv"
df = pd.read_csv(url)

# 2. 총대출액 컬럼 생성
df['총대출액'] = df['신용대출'] + df['담보대출']

# 3. 지역코드와 성별별로 총대출액 합계 계산
# groupby 후 unstack()을 하면 '성별'의 고유값들이 컬럼이 됩니다.
grouped = df.groupby(['지역코드', '성별'])['총대출액'].sum().unstack()

# 4. 성별 간 총대출액 차이가 가장 큰 지역코드 찾기
# 데이터의 성별이 1, 2라면 아래와 같이 계산합니다.
# 만약 성별이 '남', '여'라면 grouped['남'] - grouped['여']로 수정하세요.
grouped['차이'] = abs(grouped[1] - grouped[2])

# 5. 차이가 가장 큰 지역코드(인덱스) 추출
result = grouped['차이'].idxmax()

print(result)
```

```python
import pandas as pd

# 1. 데이터 로드 (URL 오타 수정: lovedlim)
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/loan.csv"
df = pd.read_csv(url)

# 2. 총대출액 컬럼 생성
df['총대출액'] = df['신용대출'] + df['담보대출']

# 3. 피벗 테이블 생성 (지역코드별, 성별별 총대출액 합계)
pivot_result = df.pivot_table(
    index='지역코드',
    columns='성별',
    values='총대출액',
    aggfunc='sum'
)

# 4. 성별 간 총대출액 차이 계산 (절대값)
# 성별 컬럼이 숫자 1, 2일 경우를 가정합니다.
# 만약 에러가 난다면 pivot_result['1'] - pivot_result['2'] 처럼 따옴표를 붙여보세요.
pivot_result['차이'] = abs(pivot_result[1] - pivot_result[2])

# 5. 차이가 가장 큰 지역코드(인덱스) 찾기
result = pivot_result['차이'].idxmax()

# 결과 출력
print(result)
```

```python
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv"
df = pd.read_csv(url)

# 2. "발생건수"와 "검거건수" 데이터 분리
df_gen = df[df["구분"] == "발생건수"].copy()
df_arr = df[df["구분"] == "검거건수"].copy()

# 3. 인덱스 초기화 (연산을 위해 행 번호를 맞춤)
df_gen = df_gen.reset_index(drop=True)
df_arr = df_arr.reset_index(drop=True)

# 4. 수치 데이터(연도별 컬럼)만 추출 (세 번째 열부터 끝까지)
gen_values = df_gen.iloc[:, 2:]
arr_values = df_arr.iloc[:, 2:]

# 5. 검거율 계산 (검거건수 / 발생건수)
df_rate = arr_values / gen_values

# 6. 각 범죄유형(행)별로 검거율이 가장 높은 '연도(열 이름)' 찾기
# axis=1을 사용하여 가로 방향(연도들 중)에서 최댓값의 컬럼명을 가져옵니다.
max_years = df_rate.idxmax(axis=1)

# 7. 해당 지점의 검거건수 합산
total_sum = 0
for idx, year in max_years.items():
    # 각 행(idx)에서 검거율이 가장 높았던 연도(year)의 검거건수를 더함
    total_sum += df_arr.loc[idx, year]

# 8. 결과 출력
print(int(total_sum)) # 출력 결과: 7799
```

```python
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv"
df = pd.read_csv(url)

# 2. 데이터를 길게 변환 (melt)
# 연도와 구분(발생/검거)을 기준으로 각 범죄유형 컬럼들을 행으로 녹입니다.
df_long = pd.melt(df, id_vars=["연도", "구분"], var_name="범죄유형", value_name="건수")

# 3. 피벗(pivot)하여 발생건수와 검거건수를 한 행에 나란히 배치
df_pivot = df_long.pivot_table(
    index=["연도", "범죄유형"],
    columns="구분",
    values="건수"
).reset_index()

# 4. 검거율 계산
df_pivot["검거율"] = df_pivot["검거건수"] / df_pivot["발생건수"]

# 5. 각 연도별로 검거율이 가장 높은 행의 인덱스(idxmax) 추출
# (주의: 변수명을 idxbox로 통일합니다)
idxbox = df_pivot.groupby("연도")["검거율"].idxmax()

# 6. 해당 인덱스에 해당하는 데이터만 추출하여 검거건수 합산
max_crime = df_pivot.loc[idxbox]
result = max_crime["검거건수"].sum()

# 7. 결과 출력
print(int(result)) # 결과: 7799
```

```python
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv"
df = pd.read_csv(url)

# 2. 만족도 결측치 처리 (전체 평균)
m_val = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m_val)

# 3. 근속연수 결측치 처리 (부서와 성과등급별 평균)
# transform을 사용해 그룹별 평균값을 각 행에 대응시킵니다.
group_mean = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")

# 평균값을 정수로 변환하여 결측치를 채웁니다. (근속연수는 보통 정수이므로)
df["근속연수"] = df["근속연수"].fillna(group_mean.astype(int))

# 4. 연봉 / 근속연수 계산 및 3번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
# 상위 3명을 뽑은 뒤, 마지막 행(즉, 3위)의 근속연수 추출
df_year = df.nlargest(3, '연봉_근속연수')
A = df_year.iloc[-1]["근속연수"]

# 5. 연봉 / 만족도 계산 및 2번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
# 상위 2명을 뽑은 뒤, 마지막 행(즉, 2위)의 교육참가횟수 추출
df_like = df.nlargest(2, '연봉_만족도')
B = df_like.iloc[-1]["교육참가횟수"]

# 6. 결과 출력
result = A + B
print(int(result))
```

```python
import pandas as pd

# 1. 데이터 로드
url = "https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv"
df = pd.read_csv(url)

# 2. 만족도 결측치 처리 (전체 평균)
m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

# 3. 근속연수 결측치 처리 (부서와 성과등급 기준 평균값으로 채움)
# transform을 사용하면 각 행에 맞는 그룹 평균값을 바로 매칭해줍니다.
gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
df["근속연수"] = df["근속연수"].fillna(gm.astype(int))

# 4. 연봉 / 근속연수 계산 및 3번째로 높은 사람의 근속연수 (A)
df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
df_year = df.sort_values(by="연봉_근속연수", ascending=False)
# 인덱스 2가 실제 3번째 행입니다.
A = df_year.iloc[2]["근속연수"]

# 5. 연봉 / 만족도 계산 및 2번째로 높은 사람의 교육참가횟수 (B)
df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.sort_values(by="연봉_만족도", ascending=False)
# 인덱스 1이 실제 2번째 행입니다.
B = df_like.iloc[1]["교육참가횟수"]

# 6. 결과 출력
result = A + B
print(int(result)) # np.int64 표시 없이 숫자만 출력
```
